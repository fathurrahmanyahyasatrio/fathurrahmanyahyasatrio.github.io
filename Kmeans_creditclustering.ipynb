We have identified correlated features in the dataset, and to address this, we will employ dimensionality reduction techniques to reduce the data to a lower-dimensional space. In this case, we will utilize Principal Component Analysis (PCA) for dimensionality reduction.

To provide a brief explanation of PCA, it seeks new dimensions or axes for the dataset that explain the maximum variance. The first principal component is determined as the axis that captures the highest amount of variance. Subsequently, additional components are selected, perpendicular to the previous ones, that explain the maximum remaining variance.

Imagine a visual representation where data points are projected onto the first principal component. If we compare this projection to other axes, the points will be more spread out along the first principal component. This indicates that the first principal component explains the highest variance and becomes the primary component. For a two-dimensional dataset, there will be only one component perpendicular to the first principal component, which becomes the second principal component.

Please note that if we had a third dimension extending out from the screen, we would have two components perpendicular to the first principal component. In such cases, we would choose the component that explains the maximum variance out of the two.

Once we have determined these principal components, we can select the desired number of components that collectively account for 95% of the variance in the original data. By expressing our data in terms of these principal components, we effectively reduce the dimensionality.

Now, let's proceed with implementing PCA by choosing the appropriate number of components that explain 95% of the variance in our original data.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
X_red = pca.fit_transform(df)
```


Now that we have completed the necessary preprocessing steps, we can finally move on to the exciting task of clustering. To extract clusters of information from our dataset, we will employ the K-means clustering algorithm.

Model Training
It is quite interesting and intuitive to know how KMeans clustering algorithm actually works.

KMeans clustering is an unsupervised clustering algorithm which groups together similar data in same cluster to form k clusters. As a result we get groups of similar records which can then be labelled and acted upon accordingly.

How does the algorithm find the clusters?

1. Given number of clusters k, it first selects k random point (may not be points in dataset) to be the k cluster centroids.

2. Then we assign each point to the closest centroid and form k clusters.

3. Once all points are assigned to a cluster, we then compute new centroids for each cluster.

4. Then we reassign points to the closest centroid.

5. If any reassignment takes place in step 4 we repeat steps 3 and 4. If no reassignment takes place then our model is ready and we have extracted k clusters from our dataset.

```python
from sklearn.cluster import KMeans

kmeans_models = [KMeans(n_clusters=k, random_state=23).fit(X_red) for k in range (1, 10)]
innertia = [model.inertia_ for model in kmeans_models]

plt.plot(range(1, 10), innertia)
plt.title('Elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

You see the elbow there? It seems like the elbow is around 3 or 4 . We will use silhouette score to see which one performs better.

```python
from sklearn.metrics import silhouette_score

silhoutte_scores = [silhouette_score(X_red, model.labels_) for model in kmeans_models[1:4]]
plt.plot(range(2,5), silhoutte_scores, "bo-")
plt.xticks([2, 3, 4])
plt.title('Silhoutte scores vs Number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutte score')
plt.show()
```

Model Evaluation and Inference.

```python
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=2, random_state=23)
kmeans.fit(X_red)

print('Silhoutte score of our model is ' + str(silhouette_score(X_red, kmeans.labels_)))
```

```python
df['cluster_id'] = kmeans.labels_
```
